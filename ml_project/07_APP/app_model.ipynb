```python
import pandas as pd
from sklearn import svm
from sklearn.model_selection import train_test_split, cross_val_score
from joblib import dump, load
```

```python
mushrooms = pd.read_csv('mushrooms.csv')
mushrooms.head(5)
```

```python
mushrooms.columns
```

```python
df = mushrooms[['class', 'cap-shape', 'cap-surface', 'cap-color', 'stalk-shape',  'ring-number', 'habitat']]
df.head(5)
```

```python
df_encoded = pd.get_dummies(df)
df_encoded.head(5)
```

```python
df_encoded.columns
```

```python
X = df_encoded.drop(['class_e', 'class_p'], axis=1)
X.head(5)
```

```python
y = df_encoded['class_e']
y.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

```python
model = svm.SVC(C=1, decision_function_shape='ovo', degree=1, gamma=1, kernel='rbf')
```

```python
model.fit(X_train, y_train)
```

```python
train_score = model.score(X_train, y_train)
train_score
```

```python
test_score = model.score(X_test, y_test)
test_score
```

```python
dump(model, 'model.joblib')
```
