# Checking correlations between different feature

## Correlation matrix

```python
correlations = {}
for feature1 in df.columns:
    for feature2 in df.columns:
        if feature1 != feature2 or not (f'{feature1} with {feature2}'in correlations.keys() or \
        f'{feature2} with {feature1}'in correlations.keys()): #protection from doubling the correlations
            le = preprocessing.LabelEncoder()
            x = pd.Series(le.fit_transform(pd.Series(df[feature1])))
            le = preprocessing.LabelEncoder()
            y = pd.Series(le.fit_transform(pd.Series(df[feature2])))
            correlations[f'{feature1} with {feature2}'] = x.corr(y)
        else:
            continue
            
#correlations do not add the value due the fact that features are qualitative variable
```
