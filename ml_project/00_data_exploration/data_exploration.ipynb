```python
import pandas as pd
```

```python
df = pd.read_csv('../mushrooms.csv')
```

```python
df.info()
```

```python
for i in df.columns:
    for ii in df[i].unique():
        print(i, ii, df[i].value_counts()[ii])       
```

```python
for i in df.columns:
    for ii in df[i].unique():
        if df[i].value_counts()[ii] < 40: 
            print(i, ii, df[i].value_counts()[ii])  
```
